# PCA - principal component Analysis

powerful technique used in data science and machine learning to reduce the dimensionality of large datasets while preserving essential patterns and trends

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import LabelEncoder
```

# Iris Dataset

The dataset contains 150 sample o Iris plants. There are 3 classes: Iris setosa, Iris Versicolour and Iris Virginica

source : https://archive.ics.uci.edu/dataset/53/iris

```python
path_file = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
iris = pd.read_csv(path_file, header=None)
iris.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
```

```python
iris.head(10)
```

```python
X = iris.iloc[:, :-1].values
y = iris.iloc[:, -1].values
```

```python
# 3 types of species
unique_values = np.unique(y)
print(unique_values)
```

```python
# reduce to 2 columns
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
print('Shape {}'.format(X_pca.shape))
```

```python
print(X_pca)
```

```python
le = LabelEncoder()
y_num = le.fit_transform(y)
print(y_num)
```

```python
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_num)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```
